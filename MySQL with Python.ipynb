```python
import mysql.connector
mydb=mysql.connector.connect(host='localhost', user='root', password='Ankur@230703')
mycursor=mydb.cursor()

mycursor.execute('CREATE DATABASE IF NOT EXISTS office')
mycursor.execute('USE office')

mycursor.execute('DROP TABLE IF EXISTS employees')

mycursor.execute('CREATE TABLE IF NOT EXISTS employees(emp_id INT PRIMARY KEY, department VARCHAR(50), salary INT)')
insert_query='INSERT INTO employees(emp_id, department, salary) VALUES(%s, %s, %s)'
values=[(1, 'HR', 20000),
       (2, 'HR', 25000),
       (3, 'IT', 40000),
       (4, 'IT', 45000),
       (5, 'Sales', 15000)]
mycursor.executemany(insert_query, values)
mydb.commit()
print('Records inserted successfully')

print('\nAll Records : ')
mycursor.execute('SELECT * FROM employees')
for row in mycursor.fetchall():
    print(row)


print('\nMaking a group of department wise : ')
mycursor.execute('SELECT department, SUM(salary) FROM employees GROUP BY department')
for row in mycursor.fetchall():
    print(row)


print('\nDepartments whose total salary is greater than 40000 : ')
mycursor.execute('SELECT department, SUM(salary) AS total_salary FROM employees GROUP BY department HAVING SUM(salary) > 40000')
for row in mycursor.fetchall():
    print(row)

print('\nDepartments having more than one employee : ')
mycursor.execute('SELECT department, COUNT(*) AS total_employee FROM employees GROUP BY department HAVING COUNT(*) > 1')
for row in mycursor.fetchall():
    print(row)


print('\nMinimum, Maximum and Average of salary : ')
mycursor.execute('SELECT MIN(salary), MAX(salary), AVG(salary) FROM employees')
for row in mycursor.fetchall():
    print(row)


mydb.close()
```

```python
# Joining two tables 

import mysql.connector

mydb=mysql.connector.connect(host='localhost', user='root', password='Ankur@230703')
mycursor=mydb.cursor()

mycursor.execute('DROP DATABASE IF EXISTS college')

mycursor.execute('CREATE DATABASE IF NOT EXISTS college')
mycursor.execute('USE college')

print('Table 1 : ')
mycursor.execute('CREATE TABLE IF NOT EXISTS students(student_id INT PRIMARY KEY, name VARCHAR(50), course_id INT)')
insert_query='INSERT INTO students(student_id, name, course_id) VALUES (%s, %s, %s)'
values=[(1, 'Ankur', 101),
       (2, 'Ravi', 102),
       (3, 'Priya', 103),
       (4, 'Aman', None)]
mycursor.executemany(insert_query, values)
mydb.commit()
mycursor.execute('SELECT * FROM students')
for row in mycursor.fetchall():
    print(row)


print('\nTable 2 : ')
mycursor.execute('CREATE TABLE IF NOT EXISTS courses(course_id INT PRIMARY KEY, course_name VARCHAR(50))')
insert_query='INSERT INTO courses(course_id, course_name) VALUES(%s, %s)'
values=[(101,'Python'),
       (102, 'SQL'),
       (104, 'AI')]
mycursor.executemany(insert_query, values)
mydb.commit()
mycursor.execute('SELECT * FROM courses')
for row in mycursor.fetchall():
    print(row)


print('\nInner Join : ')
mycursor.execute('SELECT students.name, courses.course_name FROM students INNER JOIN courses ON students.course_id=courses.course_id')
for row in mycursor.fetchall():
    print(row)

print('\nLeft Join : ')
mycursor.execute('SELECT students.name, courses.course_name FROM students LEFT JOIN courses ON students.course_id=courses.course_id')
for row in mycursor.fetchall():
    print(row)

print('\nRight Join : ')
mycursor.execute('SELECT students.name, courses.course_name FROM students RIGHT JOIN courses ON students.course_id=courses.course_id')
for row in mycursor.fetchall():
    print(row)

# print('\nFull Outer Join : ')
# mycursor.execute('SELECT students.name, courses.course_name FROM students FULL OUTER JOIN courses ON students.course_id=courses.course_id')
# for row in mycursor.fetchall():
#     print(row)

print('\nTable 3 : ')
mycursor.execute('CREATE TABLE IF NOT EXISTS employees (emp_id INT PRIMARY KEY, emp_name VARCHAR(50), manager_id INT)')
insert_query='INSERT INTO employees (emp_id, emp_name, manager_id) VALUES (%s, %s, %s)'
values=[(1, 'Raj', None),
       (2, 'Amit', 1),
       (3, 'Neha', 1),
       (4, 'Vikas', 2)]
mycursor.executemany(insert_query, values)
mydb.commit()
mycursor.execute('SELECT * FROM employees')
for row in mycursor.fetchall():
    print(row)

print('\nSelf Join : ')
mycursor.execute('SELECT e.emp_name AS Employee, m.emp_name AS Manager FROM employees e LEFT JOIN employees m ON e.manager_id=m.emp_id')
for row in mycursor.fetchall():
    print(row)
```
